# Transform

### Creating top_imdb - 1NF

```python
top_imdb = OBDM_250[['imdbID', 'Title']].copy()

# Renaming column headers
top_imdb = top_imdb.rename(columns = {'imdbID': 'imdb_id',
                                     'Title': 'movie_title'})
top_imdb
```

```python
# Saving DataFrame to csv file
top_imdb.to_csv('../Output/database/top_imdb.csv', index=False)
```

# Creating movie - 1NF

```python
# merging dataframe on "Title"
movie = pd.merge(top_250_imdb, OBDM_250, on='Title')
```

```python
movie.head(2)
```

```python
movie = movie[['imdbID', 'IMDb Rank', 'Title', 'Year', 'Runtime', 'Rated', 'imdbRating', 'Production']].copy()
movie['Year'] = movie['Year'].astype(float).map("{:.0f}".format)
```

```python
movie = movie.rename(columns = {'imdbID': 'imdb_id',
                                'IMDb Rank': "imdb_rank",
                                'Title': 'title',
                                'Year': 'year',
                                'Runtime': 'runtime',
                                'Rated': 'rated',
                                'imdbRating': 'imdb_rating',
                                'Production': 'production'})
movie['runtime'] = movie['runtime'].str.split(" ", expand=True)[0]
```

```python
movie
```

```python
# Saving DataFrame to csv file
movie.to_csv('../Output/database/movie.csv', index=False)
```

### Creating movie_actor - 1NF

```python
movie_actor = OBDM_250[['imdbID', 'Actors']]
movie_actor.head(5)
```

```python
# Transform each actor of a list-like to a row, replicating imdbID values.
movie_actor = movie_actor.assign(Actors=movie_actor['Actors'].str.split(', ')).explode('Actors')
movie_actor = movie_actor.reset_index(drop=True)
```

```python
movie_actor
```

```python
movie_actor.columns=['imdb_id', 'actor']
```

```python
movie_actor
```

```python
# Saving DataFrame to csv file
movie_actor.to_csv('../Output/database/movie_actor.csv', index=False)
```

### Creating movie_director - 1NF

```python
movie_director = OBDM_250[['imdbID', 'Director']]
movie_director.head(5)
```

```python
# Transform each director of a list-like to a row, replicating imdbID values.
movie_director = movie_director.assign(Director=movie_director['Director'].str.split(', ')).explode('Director')
movie_director = movie_director.reset_index(drop=True)
```

```python
movie_director
```

```python
movie_director.columns=['imdb_id', 'director']
```

```python
movie_director
```

```python
# Saving DataFrame to csv file
movie_director.to_csv('../Output/database/movie_director.csv', index=False)
```
