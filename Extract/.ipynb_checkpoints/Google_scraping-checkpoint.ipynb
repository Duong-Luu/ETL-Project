```python
# Import Splinter, BeautifulSoup, Regular Expression, and Time
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
import time
import pandas as pd
```

```python

top_250_imdb = pd.read_csv("../Output/top_250_imdb.csv")
top_250_imdb.head(5)
```

# Scaping Google

```python
# Path to chromedriver
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://www.google.com/search?q="
#browser.visit(url)
```

```python
movie_list = top_250_imdb["Movie Title"]

base_url = 'https://www.google.com/search?q='

streaming = []
title = []
price = []

for movie in movie_list:
    query_url = (f'{base_url}{movie}')

    browser.visit(query_url)

    time.sleep(3)

    soup = bs(browser.html, 'lxml')


    results1 = soup.find_all('div', class_ = 'ellip bclEt')

    for result in results1:
     #   try:
            streaming.append(result.text)
            title.append(movie.capitalize())
      #  except:
       #     Streaming.append('None')
       #     Title.append('None')
            
    results2 = soup.find_all('div', class_ = 'ellip rsj3fb')

    for result in results2:
     #   try:
            price.append(result.text)
      #  except:
      #      price.append('None')
    if len(streaming) & len(title) != len(price):
        streaming.pop()
        title.pop()
            
    print(f'{movie}:{len(title)}, {len(streaming)}, {len(price)}')
```
